```python
# Program 5-7

#Obtain the response of the system
num = [25]
den = [1,6, 25]
P_s = tf(num,den)
print(P_s)

T = np.linspace(0,5,1000)
T, y_step = control.step_response(P_s,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.grid(True)

#Check Rise time
tr = T[np.where(y_step>1)][0]
print('Rise Time: ' + repr(tr))

#Check Peak Time
tp = T[np.argmax(y_step)]
print('Peak Time: ' + repr(tp))

#Check Max Overshot
Mp = np.max(y_step)-1
print('Max Overshot: ' + repr(Mp))

#Check settling time
ts=0

for i in range(len(y_step) - 1,0, -1):
    if y_step[i] < 0.98 or y_step[i]>1.02:
        print('Settling Time: ' + repr(T[(i+1)]))
        break
```
