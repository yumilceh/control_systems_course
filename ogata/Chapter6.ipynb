```python
# Program 6–1
from scipy import signal as sg
import control as cnt
num = [1, 3]
den = [1, 5, 20, 16, 0]
P_s = tf(num, den);
print(P_s)

r_locus = control.root_locus(P_s);

plt.axis('square')
plt.grid(True);
plt.title("Root-Locus Plot of G(s) = K(s + 3)/[s(s + 1)(s^2 + 4s + 16)]",  
          weight = 'bold');
```
